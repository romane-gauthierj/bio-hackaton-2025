```r
library(Matrix)

# Directory containing .rds files
data_dir <- "/sharedFolder/Data"
file_list <- list.files(data_dir, pattern = "\\.rds$", full.names = TRUE)

# === First pass: find common genes (rows) across all samples
cat("ğŸ” Detecting common genes across all samples...\n")
common_genes <- NULL

for (f in file_list) {
  obj <- readRDS(f)
  if (is.list(obj) && !is.null(obj$exon)) {
    genes <- rownames(obj$exon)
    common_genes <- if (is.null(common_genes)) genes else intersect(common_genes, genes)
  }
}

# === Second pass: build the combined sparse matrix block by block
all_cells <- c()
mat_list <- list()

for (f in file_list) {
  obj <- readRDS(f)
  sample_id <- tools::file_path_sans_ext(basename(f))

  if (is.list(obj) && !is.null(obj$exon)) {
    mat <- obj$exon[common_genes, , drop = FALSE]
    colnames(mat) <- paste0(sample_id, "_", colnames(mat))
    mat_list[[sample_id]] <- mat
    all_cells <- c(all_cells, colnames(mat))
  }
}

# Combine all matrices column-wise
combined <- do.call(cbind, mat_list)

# === Write output in Matrix Market format (.mtx)
output_dir <- "/sharedFolder/Data/mtx_output"
dir.create(output_dir, showWarnings = FALSE)

writeMM(combined, file.path(output_dir, "matrix.mtx"))
write.table(common_genes, file.path(output_dir, "features.tsv"),
            row.names = FALSE, col.names = FALSE, quote = FALSE)
write.table(all_cells, file.path(output_dir, "barcodes.tsv"),
            row.names = FALSE, col.names = FALSE, quote = FALSE)

cat("âœ… Combined sparse matrix saved to:", output_dir, "\n")
```

```r
library(Matrix)
library(stats)
library(ggplot2)

# === Load matrix and metadata ===
matrix_file <- "/sharedFolder/Data/mtx_output/matrix.mtx"
features_file <- "/sharedFolder/Data/mtx_output/features.tsv"
barcodes_file <- "/sharedFolder/Data/mtx_output/barcodes.tsv"

mat <- readMM(matrix_file)
genes <- readLines(features_file)
barcodes <- readLines(barcodes_file)

# Apply row and column names
rownames(mat) <- genes
colnames(mat) <- barcodes

cat("âœ… Loaded matrix: ", nrow(mat), " genes x ", ncol(mat), " cells\n")

# === Subsample 500 cells ===
set.seed(42)
selected_cells <- sample(colnames(mat), 500)
mat_sub <- mat[, selected_cells]

# === Select top 1000 most variable genes ===
gene_vars <- apply(mat_sub, 1, var)
top_genes <- names(sort(gene_vars, decreasing = TRUE))[1:1000]
mat_filtered <- mat_sub[top_genes, ]

# === PCA on transposed matrix (cells Ã— genes) ===
mat_dense <- as.matrix(t(mat_filtered))
mat_scaled <- scale(mat_dense)
pca_res <- prcomp(mat_scaled, rank. = 10)
pca_df <- as.data.frame(pca_res$x[, 1:2])
pca_df$cell <- rownames(pca_df)

# === K-means clustering ===
set.seed(42)
k <- 5
kres <- kmeans(pca_df[, 1:2], centers = k)
pca_df$kmeans_cluster <- factor(kres$cluster)

# === Hierarchical clustering ===
d <- dist(pca_df[, 1:2])
hc <- hclust(d, method = "ward.D2")
pca_df$hclust_cluster <- factor(cutree(hc, k = k))

# === Plotting ===
p1 <- ggplot(pca_df, aes(PC1, PC2, color = kmeans_cluster)) +
  geom_point(alpha = 0.7) + theme_minimal() + ggtitle("K-means clustering")

p2 <- ggplot(pca_df, aes(PC1, PC2, color = hclust_cluster)) +
  geom_point(alpha = 0.7) + theme_minimal() + ggtitle("Hierarchical clustering")

print(p1)
print(p2)
```
